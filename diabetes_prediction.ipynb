##Diabates Prediction
###Training a Suport Vector Machine

Importing libraries

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

###Data Collection and Analysis

```python
df = pd.read_csv('/content/drive/MyDrive/datasets/diabetes.csv')
```

```python
df.head()
```

```python
#number of rows and columns in the dataframe
df.shape
```

```python
#statistical measure of the data
df.describe()
```

```python
# 0 means non diabetic, and 1 means diabetic
df['Outcome'].value_counts()
```

```python
df.groupby('Outcome').mean()
```

```python
X = df.drop(columns = 'Outcome', axis=1)
Y = df['Outcome']
```

###Data Standardization

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

###Training the model

```python
model = svm.SVC(kernel = 'linear')
```

```python
model.fit(X_train, Y_train)
```

###Model evaluation

```python
prediction = model.predict(X_test)
pred_accuracy = accuracy_score(prediction, Y_test)
print("Accuracy score: ", pred_accuracy)
```

###Building a predictive system

```python
# taking a sample from dataset: 1,189,60,23,846,30.1,0.398,59 -> 1 = diabetics
input_data = (1,189,60,23,846,30.1,0.398,59)

#input_data to numpy array and reshape it
input_data = np.asarray(input_data).reshape(1, -1)

pred = model.predict(input_data)
if pred[0] == 0:
  print('The person isn\'t diabetics')
else:
  print ('The person is diabetics')
```

### Saving the model

```python
import pickle
```

```python
filename = 'diabetes_model.sav'
pickle.dump(model, open(filename, 'wb'))
```
