### Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/content/drive/MyDrive/datasets/creditcard.csv')
```

#### Under-sampling:
Builds a dataset containing similar ditribution for Legit and Fraud transactions

```python
legit_sample = legit.sample(n=492)
```

```python
balanced_df = pd.concat([legit_sample, fraud], axis=0)
```

```python
balanced_df.shape
```

```python
balanced_df['Class'].value_counts()
```

#### Splitting the data

```python
X = balanced_df.drop(columns='Class', axis=1)
Y = balanced_df['Class']
print('X: {}\nY: {}'.format(X.shape, Y.shape))
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, stratify=Y)
```

```python
print('''
x_train.shape: {}
x_test.shape: {}
y_train.shape: {}
y_test.shape: {}
'''.format(x_train.shape, x_test.shape,
          y_train.shape, y_test.shape))
```

### Training the model

####Logistic Regression

```python
log_reg_model = LogisticRegression(max_iter=110)
```

```python
log_reg_model.fit(x_train, y_train)
```

### Evaluating the model

```python
# Logistic Regression
log_reg_pred = log_reg_model.predict(x_test)
acc = accuracy_score(y_test, log_reg_pred)
print('Accuracy score: ', acc)
```
